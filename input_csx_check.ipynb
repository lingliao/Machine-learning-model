```r
install.packages("tidyverse")
library(tidyverse)
```

```r
p3 <-  ggplot(input,aes(x=data, y=Fraction)) +
  geom_violin(trim=FALSE) +
  geom_dotplot(binaxis='y', stackdir='center', ,dotsize=1, fill = "#E69F00",colour="#E69F00")+
  stat_summary(fun.y = mean, aes(colour="mean", shape="mean"), 
               size = 1) +
  stat_summary(fun.y =median, aes(colour="median", shape="median"), 
               size=4, geom="point") +
  labs(title = paste0("test"), x = "CSx output", y = "Fraction", colour=NULL, shape=NULL) +
  theme_classic() +
  scale_colour_manual(values=c("blue", "red")) +
  scale_shape_manual(values=c(16, 5))
p3

ggsave(filename = "CSx_data.pdf",width=8, height=8, dpi = 700)
```
